#**Exercise 5: Basic Data Analysis with Google Colab**

```python
df_iris = pd.read_csv("Iris.csv")
df_iris.head()
#print(df_iris.describe())
#print(df_iris.info())
mean_df_iris = df_iris[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].mean()
print("\nThe mean of Quantitative columns are:\n ", mean_df_iris)
median_df_iris = df_iris[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].median()
print("\nThe median of Quantitative columns are:\n ",median_df_iris)
mode_df_iris  = df_iris['Species'].mode()
print("\nThe mode of Qualitative columns is:\n ",mode_df_iris)
```

```python
import matplotlib.pyplot as plt
#plt.plot(df_iris['SepalLengthCm'][1:5], df_iris['PetalLengthCm'][1:5])
plt.plot(df_iris['SepalLengthCm'], df_iris['PetalLengthCm'])
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot Sepal Length
axs[0,0].plot(df_iris['SepalLengthCm'], color='blue')
axs[0,0].set_title("Sepal Length (cm)")

# Plot Sepal Width
axs[0,1].plot(df_iris['SepalWidthCm'], color='green')
axs[0,1].set_title("Sepal Width (cm)")

# Plot Petal Length
axs[1,0].plot(df_iris['PetalLengthCm'], color='red')
axs[1,0].set_title("Petal Length (cm)")

# Plot Petal Width
axs[1,1].plot(df_iris['PetalWidthCm'], color='purple')
axs[1,1].set_title("Petal Width (cm)")

# Adjust layout so titles don’t overlap
plt.tight_layout()
plt.show()
```
