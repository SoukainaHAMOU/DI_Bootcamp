#**🌟 Exercise 2: Creating a Line Plot for Temperature Variation**


```python
import numpy as np
import matplotlib.pyplot as plt

temperature = [72, 74, 76, 80, 82, 78, 75]
days = np.arange(len(temperature))
plt.plot(days, temperature, marker='o', linestyle='-', color='b')
plt.xlabel('Days', fontsize=16)
plt.ylabel('Temperature (°F)', fontsize=16)
plt.title('Temperature Variation Over Days')
plt.xticks(days,['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.grid(True)
plt.show()
```

#**Exercise 3: Visualizing Monthly Sales with a Bar Chart**

```python
import numpy as np
import matplotlib.pyplot as plt

sales = [5000, 5500, 6200, 7000, 7500]
months = np.arange(len(sales))

plt.bar(months, sales, color='g')
plt.xlabel('Months', fontsize=16)
plt.ylabel('Sales Amount ($)', fontsize=16)
plt.xticks(months, ['Month1', 'Month2', 'Month3', 'Month4', 'Month5'])
plt.title('Monthly Sales')
```

#**Exercise 5: Comparing Anxiety Levels Across Different Genders**


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('Student Mental health.csv')
df
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Do you have Anxiety?', hue='Choose your gender')
plt.xlabel('Anxiety Level')
plt.ylabel('Number of Students')
plt.title('Comparison of Anxiety Levels Across Different Genders')

plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('Student Mental health.csv')

df['Do you have Panic attack?'] = df['Do you have Panic attack?'].map({'Yes': 1, 'No': 0})
# df['Do you have Panic attack?'] = df['Do you have Panic attack?'].replace({'Yes': 1, 'No': 0})

sns.scatterplot(data=df, x='Age', y='Do you have Panic attack?',hue='Choose your gender',style='Choose your gender',s=80,alpha=0.7)

plt.xlabel('Age of Students', fontsize=12)
plt.ylabel('Panic Attack (Yes=1, No=0)', fontsize=12)
plt.title('Relationship Between Age and Panic Attacks', fontsize=14, fontweight='bold')
plt.show()
```
