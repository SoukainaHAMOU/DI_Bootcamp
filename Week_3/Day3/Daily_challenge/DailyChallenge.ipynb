```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['salary_norma'] = scaler.fit_transform(df[['salary']])
df['salary_norma'].describe()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE


# Drop target column (salary) and keep only numeric features
X = df.drop('salary', axis=1).select_dtypes(include=['float64', 'int64'])

# Scale features before PCA/t-SNE (important for fair comparison)
std_scaler = StandardScaler()
X_scaled = std_scaler.fit_transform(X)

# PCA ---
pca = PCA(n_components=2)
df_pca = pca.fit_transform(X_scaled)
df_pca = pd.DataFrame(df_pca, columns=['PC1', 'PC2'])

#  t-SNE ---
tsne = TSNE(n_components=2, random_state=42)
df_tsne = tsne.fit_transform(X_scaled)
df_tsne = pd.DataFrame(df_tsne, columns=['TSNE1', 'TSNE2'])
```

Group the dataset by the 'experience_level' column and calculate the average and median salary for each experience level (e.g., Junior, Mid-level, Senior)

```python
X_grouped = df.groupby('experience_level')

df_grouped_stats = X_grouped['salary'].agg(['mean', 'median', 'max', 'min'])

print(df_grouped_stats)
```
