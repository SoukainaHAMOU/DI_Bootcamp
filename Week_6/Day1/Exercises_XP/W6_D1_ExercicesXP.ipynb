```python

# This piece of code is already prefilled, run it to execute it and see the results.
# It demonstrates standard classification metrics for binary loan default prediction.

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, average_precision_score, confusion_matrix, classification_report

# Please replace these placeholders with your true labels and predicted probabilities.
y_true = [0,1,0,1,0,0,1,0,1,0]            # placeholder labels
y_pred_proba = [0.05,0.80,0.10,0.65,0.20,0.15,0.70,0.30,0.85,0.25]  # placeholder probabilities

# You should set a decision threshold that reflects the precision–recall trade-off for your business case.
threshold = 0.5
y_pred = [1 if p >= threshold else 0 for p in y_pred_proba]

print("Accuracy:", round(accuracy_score(y_true, y_pred), 4))
print("Precision:", round(precision_score(y_true, y_pred, zero_division=0), 4))
print("Recall:", round(recall_score(y_true, y_pred, zero_division=0), 4))
print("F1-score:", round(f1_score(y_true, y_pred, zero_division=0), 4))
print("ROC-AUC:", round(roc_auc_score(y_true, y_pred_proba), 4))
print("PR-AUC (Average Precision):", round(average_precision_score(y_true, y_pred_proba), 4))
print("\nConfusion matrix:\n", confusion_matrix(y_true, y_pred))
print("\nClassification report:\n", classification_report(y_true, y_pred, zero_division=0))
```

### Answer
> For loan default prediction, I would apply Logistic Regression because it is a simple classification of two classes.

**Evaluation Plan**
> I would split the dataset into training, and test sets.

```python

# This piece of code is already prefilled, run it to execute it and see the results.
# Unsupervised clustering metrics template with synthetic data.

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

X, _ = make_blobs(n_samples=300, centers=3, random_state=42)
kmeans = KMeans(n_clusters=3, n_init="auto", random_state=42)
labels = kmeans.fit_predict(X)
sil = silhouette_score(X, labels)
print("Silhouette score (higher is better):", round(sil, 4))

print("Please explain in complete sentences when you would use the elbow method and how you would interpret it.")
```

### Answer

**Supervised Learning**

For a supervised classification model such as Logistic Regression or Random Forest, I would evaluate performance using a train/test split combined with stratified k-fold cross-validation to ensure balanced representation of classes. I would report metrics including accuracy, precision, recall, F1-score, ROC-AUC, and PR-AUC, since these capture both overall correctness and the trade-off between false positives and false negatives. I would also use ROC curves and precision–recall curves to visualize threshold effects and select a decision threshold that aligns with business priorities.

**Unsupervised Learning**

For an unsupervised clustering model such as K-Means, I would measure cluster quality using the silhouette score, which evaluates how similar each point is to its own cluster compared to other clusters. I would also apply the elbow method, plotting the within-cluster sum of squares against the number of clusters, and interpret the “elbow point” as the optimal number of clusters where adding more clusters no longer significantly reduces variance.  Diagnostic plots such as cluster scatterplots would help visualize the structure and separation of clusters. A key challenge in unsupervised evaluation is the absence of ground truth labels, which makes it difficult to judge whether clusters are truly meaningful.

**Reinforcement Learning**

For a reinforcement learning model such as Q-Learning or Deep Q-Networks, I would measure success by tracking cumulative reward across episodes, ensuring the agent learns to maximize long-term gains. I would assess convergence by monitoring whether the policy stabilizes and performance plateaus after sufficient training.  A key challenge in reinforcement learning evaluation is that performance can vary widely across environments, and convergence may be slow or unstable depending on reward design and exploration strategy.
