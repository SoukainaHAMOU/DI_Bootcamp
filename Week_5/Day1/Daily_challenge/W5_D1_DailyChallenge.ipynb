```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#**Data Preparation:**

```python
# np.random.seed(42)
data = np.random.uniform(low=-5, high=35, size=(10, 12))

cities = [f"City_{i+1}" for i in range(10)]
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

df = pd.DataFrame(data, index=cities, columns=months)

df.head()
```

#**Data Analysis:**

```python
for city in cities:
  print(f" For City: {city}")
  mean_temperature = df.loc[city].mean()
  print(f"Mean temperature: {mean_temperature:.2f}")

  max_temperature = df.loc[city].max()
  print(f"Maximum temperature: {max_temperature:.2f}")

  min_temperature = df.loc[city].min()
  print(f"Minimum temperature: {min_temperature:.2f}")
  print("-"*30)
```

```python
# Annual average temperature per city
annual_avg = df.mean(axis=1)

# Identify hottest and coldest cities
hottest_city = annual_avg.idxmax()
coldest_city = annual_avg.idxmin()

print("Annual Average Temperatures:\n", annual_avg)
print(f"\nHottest City: {hottest_city} ({annual_avg[hottest_city]:.2f} °C)")
print(f"Coldest City: {coldest_city} ({annual_avg[coldest_city]:.2f} °C)")
```

```python
plt.figure(figsize=(20, 10))

for city in cities:
  plt.plot(df.loc[city], linestyle='-', label=city)

plt.legend()

plt.title("Monthly Temperature Trends by City")
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
plt.legend(loc="upper right")
plt.show()
```

#**Brief report**

We generated synthetic average monthly temperatures for 10 cities, ranging between -5°C and 35°C.

**Findings:**
- The hottest city of the year is **City_4** with an annual average of 23.30°C.
- The coldest city of the year is **City_2** with an annual average of 10.75°C.


**Interesting Trends:**
- Cities with higher averages tend to peak in June until August.
- Cooler cities show more variation across months.
