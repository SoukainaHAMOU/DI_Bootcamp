#**🌟 Exercise 6 : Basic Hypothesis Testing**

```python
# Productivity scores of employees before the training program
productivity_before = np.random.normal(loc=50, scale=10, size=30)

# Productivity scores of the same employees after the training program
productivity_after = productivity_before + np.random.normal(loc=5, scale=3, size=30)

# Your task is to formulate a hypothesis regarding the training program's effectiveness
# and test it using basic statistical functions in NumPy.

# Differences between before and after
differences = productivity_after - productivity_before

# Mean and standard deviation of differences
mean_diff = np.mean(differences)
std_diff = np.std(differences, ddof=1)

# Compute a simple test statistic (like a t-score)
t_stat = mean_diff / (std_diff / np.sqrt(len(differences)))

print("Mean difference:", f'{mean_diff:.2f}')
print("Standard deviation of differences:", f'{std_diff:.2f}')
print("Test statistic:", f'{t_stat:.2f}')
```

#**🌟 Exercise 8 : Time Series Data Manipulation**

```python
import pandas as pd
import numpy as np

dates = pd.date_range(start="2023-01", end="2024-01", freq="D")

values = np.random.randint(90, 110, size=len(dates))

time_series = pd.DataFrame({"Date": dates, "Value": values})
time_series.set_index("Date", inplace=True)

q1 = time_series["2023-01-01":"2023-03-31"]
q2 = time_series["2023-04-01":"2023-06-30"]
q3 = time_series["2023-07-01":"2023-09-30"]
q4 = time_series["2023-10-01":"2023-12-31"]

print("Q1 (Jan–Mar):")
print(q1.head(), "\n")   # show first few rows

print("Q2 (Apr–Jun):")
print(q2.head(), "\n")

print("Q3 (Jul–Sep):")
print(q3.head(), "\n")

print("Q4 (Oct–Dec):")
print(q4.head(), "\n")
```

#**🌟 Exercise 9 : Data Conversion**

```python
import numpy as np
import pandas as pd

numpy_array = np.array([[10, 20, 30], [40, 50, 60], [70, 80, 90]])
print("Original NumPy Array:\n", numpy_array)

# From NumPy array to Pandas DataFrame
df_from_numpy = pd.DataFrame(numpy_array, columns=['Col A', 'Col B', 'Col C'])
print("\nDataFrame from NumPy Array:\n", df_from_numpy)

# From Pandas DataFrame to NumPy array
numpy_from_df = df_from_numpy.to_numpy()
print("\nNumPy Array from DataFrame:\n", numpy_from_df)
```
