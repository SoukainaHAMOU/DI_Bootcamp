```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#**Exercise 1 : Visualizing the Distribution of CGPA**

```python
df = pd.read_csv('Student Mental health.csv')
df.head()
coolumns = df.columns
print(coolumns)
df['What is your CGPA?'].head()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='What is your CGPA?', discrete=True, shrink=0.8, color='skyblue')
plt.title('Distribution of CGPA Categories')
plt.xlabel('CGPA Category')
plt.ylabel('Count')
plt.show()
```

#**Exercise 2: Comparing Anxiety Levels Across Different Genders**

```python
sns.barplot(data=df, x= 'Choose your gender', y='Do you have Anxiety?', color='blue')

plt.title('Anxiety Levels Across Different Genders', fontsize=14, fontweight='bold')
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Average Anxiety Level', fontsize=12)

plt.show()
```

#**Exercise 3: Exploring the Relationship Between Age and Panic Attacks**

```python
df['Do you have Panic attack?'] = df['Do you have Panic attack?'].map({'Yes': 1, 'No': 0})
df.head()
sns.scatterplot(data=df, x='Age', y='Do you have Panic attack?', color='red')
plt.title('Relationship Between Age and Panic Attacks')
plt.xlabel('Age')
plt.ylabel('Panic Attacks')
plt.show()
```

#**Exercise 5: Creating a Heatmap to Visualize Correlations**

```python
subset_cols = ['Age', 'Do you have Panic attack?', 'Marital status','Do you have Depression?','Do you have Anxiety?','Did you seek any specialist for a treatment?']
df_subset = df[subset_cols]
correlation_matrix = df_subset.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of Mental Health Factors', fontsize=16)
plt.show()
```

#**🌟 Exercise 6: Analyzing Distributions and Relationships Using FacetGrid**

```python
g = sns.FacetGrid(df, col='Do you have Depression?', height=4, aspect=1.2)
g.map(sns.histplot, 'What is your CGPA?', bins=20, kde=True, color='skyblue')
g.fig.suptitle('Distribution of CGPA by Depression Status', fontsize=16, fontweight='bold', y=1.05)
g.set_axis_labels('CGPA', 'Count')
plt.show()
```
