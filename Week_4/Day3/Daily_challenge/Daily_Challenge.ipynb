```python
import numpy as np
from  matplotlib import pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
import random
import os
```

```python
from scipy.ndimage import rotate
def rotate_image_30_degrees(image):
  return rotate(image, 90, axes=(1,0), reshape=True)
```

```python
list_img = []
plt.figure(figsize=(20,20))
for i in range(5):
  file = random.choice(os.listdir('flower-color-images/flower-color-images/flowers/flowers'))
  image_path= os.path.join('flower-color-images/flower-color-images/flowers/flowers', file)
  img=mpimg.imread(image_path)
  list_img.append(img)

for i, img in enumerate(list_img):
  ax=plt.subplot(3,5,i+1)
  ax.title.set_text(file)
  plt.imshow(img)
```

```python
# To visualize the rotated images :
rotated_list = []
plt.figure(figsize=(20,20))
for img in list_img:
  rotated_list.append(rotate_image_30_degrees(img))

for i, img in enumerate(rotated_list):
    ax=plt.subplot(3,5,i+1)
    ax.title.set_text(file)
    plt.imshow(img)
```
