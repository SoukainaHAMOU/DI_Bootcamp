#**Exercise 2 : Creating a Line Plot for Temperature Variation**

```python
import matplotlib.pyplot as plt
import numpy as np

Days = np.arange(1, 8)
temperature = [72, 74, 76, 80, 82, 78, 75]
plt.plot(Days, temperature)
plt.xlabel('Days', fontsize='x-large')
plt.ylabel('Temperature (°F)”', fontsize='x-large')
plt.xticks(Days, ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"])
plt.title('Temperature Variation Over Days', fontweight='bold')
plt.show()
```

#**Exercise 3 : Visualizing Monthly Sales with a Bar Chart**

```python
#months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
months = np.arange(1, 6)
sales_amount = [5000, 5500, 6200, 7000, 7500]
plt.bar(months, sales_amount)
plt.xlabel('Months', fontsize='x-large')
plt.ylabel('Sales Amount', fontsize='x-large')
plt.xticks(months, ["Jan", "Feb", "Mar", "Apr", "May"])
plt.title('Monthly Sales', fontweight='bold')
plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('sales_data.csv')
df

total_quantity = df['quantity'].sum()
print("Total Quantity of Products Sold:", total_quantity)


category_revenue = df.groupby('category')['revenue'].sum()
highest_category = category_revenue.idxmax()
highest_revenue = category_revenue.max()
print("Category with Highest Revenue:", highest_category)
print("Revenue Generated by the Category:", highest_revenue)

average_revenue = df['revenue'].mean()
print(f"Average Revenue per Sale: {average_revenue:.2f}")

df['date'] = pd.to_datetime(df['date'])
df['quarter'] = df['date'].dt.quarter
quarterly_revenue = df.groupby('quarter')['revenue'].sum()

plt.figure(figsize=(8,5))
plt.bar(quarterly_revenue.index, quarterly_revenue.values, color='skyblue')
plt.xticks([1,2,3,4], ['Q1','Q2','Q3','Q4'])
plt.xlabel('Quarter')
plt.ylabel('Total Revenue')
plt.title('Total Revenue by Quarter')
plt.show()


# plt.figure(figsize=(8,5))
# category_revenue.plot(kind='bar', color='lightgreen')
# plt.xlabel('Category')
# plt.ylabel('Total Revenue')
# plt.title('Total Revenue by Category')
# plt.show()
```

#**Data Visualisation Using MatPlotLib**
Create a simple line plot using Matplotlib. Plot the function y = x^2 for x values ranging from -10 to 10. Add a title and label the x and y axes.

Generate a bar chart to display the sales data of four different products (A, B, C, D) in a week. The sales values are 15, 30, 45, and 20 respectively. Label the chart appropriately.

Create a pie chart representing the following data about favorite fruits of a group of people: Apples (40%), Bananas (30%), Cherries (20%), and Dates (10%). Customize the chart with different colors for each fruit and add a legend.

```python
import matplotlib.pyplot as plt
import numpy as np

#1
x = np.arange(-10, 11)
y = x**2
plt.plot(x, y)
plt.title('Function: y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

#2
products = np.arange(1,5)
sales = [15, 30, 45, 20]
plt.bar(products, sales)
plt.xlabel('Products')
plt.ylabel('Sales')
plt.xticks(products, ['A', 'B', 'C', 'D'])
plt.title('Sales Data for Four Products')
plt.show()

#3
fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
favorite = [40, 30, 20, 10]
colors = ['green', 'yellow', 'red', 'brown']
plt.pie(favorite, labels=fruits, colors=colors, autopct='%1.1f%%')
plt.title('Favorite Fruits Distribution')
plt.show()
#
```
